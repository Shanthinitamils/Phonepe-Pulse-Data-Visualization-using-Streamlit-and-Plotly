```python
# Necessary import Funcyions
import pandas as pd
import json
import os
import git
import mysql.connector
import plotly.io as pio
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Get a data from Phonepe data for aggregated transaction using GIT Clone
# Path for acces data Folder which us dowload and stored in local
def aggregate_trans():

    path="D:/PHONEPAY/data/data/aggregated/transaction/country/india/state/"
    Agg_state_list=os.listdir(path)
    Agg_state_list

    # Dictionary to append data which we iterate from json file
    clm={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

    # Iter data from aggregate trans Json file

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
                for z in D['data']['transactionData']:
                    Name=z['name']
                    count=z['paymentInstruments'][0]['count']
                    amount=z['paymentInstruments'][0]['amount']
                    clm['Transaction_type'].append(Name)
                    clm['Transaction_count'].append(count)
                    clm['Transaction_amount'].append(amount)
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quarter'].append(int(k.strip('.json')))

        #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(clm)

        # Data Manipulation

    Agg_Trans['State']=Agg_Trans['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
    Agg_Trans['State']=Agg_Trans['State'].str.replace('-'," ")
    Agg_Trans['State']=Agg_Trans['State'].str.title()
    Agg_Trans['State']=Agg_Trans['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')

    return Agg_Trans
```

```python
# Get a data from Phonepe data for aggregated User using GIT Clone
def agg_user():
    path1="D:/PHONEPAY/data/data/aggregated/user/country/india/state/"
    Agg_state_userlist=os.listdir(path1)
    Agg_state_userlist

    # Dictionary to append data which we iterate from json file

    clm1={'State':[], 'Year':[],'Quarter':[],'Brand':[],'Total_count':[],'Percentage':[]}

    for i in Agg_state_userlist:
        p_i=path1+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                E=json.load(Data)
                try:
                    for z in E['data']['usersByDevice']:
                        brand=z['brand']
                        count=z['count']
                        percentage=z["percentage"]
                        clm1['Brand'].append(brand)
                        clm1['Total_count'].append(count)
                        clm1['Percentage'].append(percentage)
                        clm1['State'].append(i)
                        clm1['Year'].append(j)
                        clm1['Quarter'].append(int(k.strip('.json')))
                except:
                    pass
    #Succesfully created a dataframe
    Agg_user=pd.DataFrame(clm1)

    # Data Manipulation

    Agg_user['State']=Agg_user['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
    Agg_user['State']=Agg_user['State'].str.replace('-'," ")
    Agg_user['State']=Agg_user['State'].str.title()
    Agg_user['State']=Agg_user['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
    return Agg_user
```

```python
def map_trans():
    path2="D:/PHONEPAY/data/data/map/transaction/hover/country/india/state/"
    Map_state_list=os.listdir(path2)
    Map_state_list



    # Dictionary to append data which we iterate from json file

    clm2={'State':[], 'Year':[],'Quarter':[],'Districts':[],'Total_count':[],'Total_Amount':[]}

    # Iter data from aggregate trans Json file

    for i in Map_state_list:
        p_i=path2+i+"/"
        Map_yr=os.listdir(p_i)
        for j in Map_yr:
            p_j=p_i+j+"/"
            Map_yr_list=os.listdir(p_j)
            for k in Map_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                F=json.load(Data)
                for z in F['data']['hoverDataList']:
                        name=z['name']
                        count=z['metric'][0]['count']
                        amount=z['metric'][0]['amount']
                        clm2['Districts'].append(name)
                        clm2['Total_count'].append(count)
                        clm2['Total_Amount'].append(amount)
                        clm2['State'].append(i)
                        clm2['Year'].append(j)
                        clm2['Quarter'].append(int(k.strip('.json')))

    Map_trans=pd.DataFrame(clm2)

    # Data Manipulation

    Map_trans['State']=Map_trans['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
    Map_trans['State']=Map_trans['State'].str.replace('-'," ")
    Map_trans['State']=Map_trans['State'].str.title()
    Map_trans['State']=Map_trans['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
    return Map_trans
```

```python
def map_user():
    path3="D:/PHONEPAY/data/data/map/user/hover/country/india/state/"
    Map_user_list=os.listdir(path3)
    Map_user_list

    # Dictionary to append data which we iterate from json file

    clm3={'State':[], 'Year':[],'Quarter':[],'Districts':[],'Register_user':[],'App_opened':[]}

    # Iter data from aggregate trans Json file

    for i in Map_user_list:
        p_i=path3+i+"/"
        Map_yr=os.listdir(p_i)
        for j in Map_yr:
            p_j=p_i+j+"/"
            Map_yr_list=os.listdir(p_j)
            for k in Map_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                G=json.load(Data)

                for z in G['data']['hoverData'].items():
                    District=z[0]
                    registeruser=z[1]['registeredUsers']
                    appuser=z[1]['appOpens']
                    clm3['Districts'].append(District)
                    clm3['Register_user'].append(registeruser)
                    clm3['App_opened'].append(appuser)
                    clm3['State'].append(i)
                    clm3['Year'].append(j)
                    clm3['Quarter'].append(int(k.strip('.json')))

    Map_user=pd.DataFrame(clm3)

    # Data Manipulation

    Map_user['State']=Map_user['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
    Map_user['State']=Map_user['State'].str.replace('-'," ")
    Map_user['State']=Map_user['State'].str.title()
    Map_user['State']=Map_user['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
    return Map_user
```

```python
def top_trans():
    path4="D:/PHONEPAY/data/data/top/transaction/country/india/state/"
    Top_trans_list=os.listdir(path4)
    Top_trans_list


    # Dictionary to append data which we iterate from json file

    clm4={'State':[], 'Year':[],'Quarter':[],'Pincode':[],'Total_count':[],'Total_Amount':[]}


    # Iter data from aggregate trans Json file

    for i in Top_trans_list:
        p_i=path4+i+"/"
        Top_yr=os.listdir(p_i)
        for j in Top_yr:
            p_j=p_i+j+"/"
            Top_yr_list=os.listdir(p_j)
            for k in Top_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                H=json.load(Data)
                for z in H['data']['pincodes']:
                    name=z['entityName']
                    count=z['metric']['count']
                    amount=z['metric']['amount']
                    clm4['Pincode'].append(name)
                    clm4['Total_count'].append(count)
                    clm4['Total_Amount'].append(amount)
                    clm4['State'].append(i)
                    clm4['Year'].append(j)
                    clm4['Quarter'].append(int(k.strip('.json')))

    Top_trans=pd.DataFrame(clm4)

    # Data Manipulation

    Top_trans['State']=Top_trans['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
    Top_trans['State']=Top_trans['State'].str.replace('-'," ")
    Top_trans['State']=Top_trans['State'].str.title()
    Top_trans['State']=Top_trans['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
    return Top_trans
```

```python
# create a database and table (Aggregate trasaction)
mycursor.execute('create database if not exists phonepe_data')
mycursor.execute('USE phonepe_data')
mycursor.execute('''create table if not exists aggregate_trans(State VARCHAR(255),Year INT(10),
                 Quarter INT(10),Transaction_type VARCHAR(255),
                 Transaction_count INT(10),Transaction_amount FLOAT(50))''')
for index,row in aggregate_trans().iterrows():
    mycursor.execute('''Insert into aggregate_trans(State,Year,Quarter,Transaction_type,Transaction_count,Transaction_amount)
                     VALUES(%s,%s,%s,%s,%s,%s)''',
                     (row['State'],row['Year'],row['Quarter'],row['Transaction_type'],row['Transaction_count'],row['Transaction_amount']))
    
mydb.commit()
```
